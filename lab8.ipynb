```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_max = 20
x = np.linspace(-np.pi, np.pi, 1000)
f = lambda x: x
n = 0
a_0 = lambda f, x, n: np.trapz((1/(np.pi))*f(x), x, dx=0.01)
a_n = lambda f, x, n: np.trapz((1/np.pi)*f(x)*np.cos(n*x), x, dx=0.01)
b_n = lambda f, x, n: np.trapz((1/np.pi)*f(x)*np.sin(n*x), x,  dx=0.01)
approx = (a_0(f, x, i))/2

for i in range(1, n_max):
    approx+=a_n(f, x, i)*np.cos(i*x) + b_n(f, x, i)*np.sin(i*x)
    plt.show(block=False)
    plt.plot(x, approx)
    plt.ylim((-np.pi, np.pi))
    plt.xlim((-np.pi, np.pi))
    plt.plot(x, f(x))
    plt.show()
```
