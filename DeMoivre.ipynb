```python
import numpy as np
import matplotlib.pyplot as plt
```

Calculating n roots of complex number z

```python
def deMoivre(z, n):
    real = lambda n,k,z: np.cos((np.angle(z)+2*np.pi*k)/n)  # finds real coefficient of k'th root
    imag = lambda n,k,z: np.sin((np.angle(z)+2*np.pi*k)/n)  # finds imaginary coefficient of k'th root
           
    
    real_coeffs = np.fromiter( ( (real(n,k,z)) for k in range(n)), dtype='float' )  # generate all real
    imag_coeffs = np.fromiter( ( (imag(n,k,z)) for k in range(n)), dtype='float' )  # and imag. coefficients
    
    #find the roots
    res = [ ( np.abs(z)**(1/n) )*(real_coeffs[k] + 1j*imag_coeffs[k]) for k in range(0,n) ]
    
    return res

deMoivre(1+1j, 4)
```

Display the roots with polar plot

```python
def main():
    data = deMoivre(7 +5j, 5)
    theta = np.angle(data)
    r = np.abs(data)
    z = plt.subplot(111, polar=True)
    for i in data:
        plt.polar(theta, r, marker='X')
        
main()
        
    
        
```
